```python
import pandas as pd;
import numpy as np;
from statistics import mean, stdev;
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("starter/files/student-mat.csv")
```

```python
data.head()
```

```python
data.isnull()
```

```python
data.dtypes
```

```python
pd.options.display.max_columns = None
display(data)
```

```python
G3_mean = data['G3'].mean()
G2_mean = data['G2'].mean()
G1_mean = data['G1'].mean()
G3_stdev = data['G3'].std()
G2_stdev = data['G2'].std()
G1_stdev = data['G1'].std()
```

```python
plt.bar(['G3','G2', 'G1'],[G3_mean, G2_mean, G1_mean])
plt.show()
```

```python
plt.bar(['G3','G2', 'G1'],[G3_mean, G2_mean, G1_mean])
plt.xlabel('Grade')
plt.ylabel('Academic Performance')
plt.title('Academic Performance of Students in Three Grades')
plt.show()
```

```python
plt.bar(['G3','G2','G1'],[G3_mean,G2_mean,G1_mean],yerr=[G3_stdev, G2_stdev, G1_stdev],capsize=5)
plt.xlabel('Grade')
plt.ylabel('Academic Performance')
plt.title('Academic Performance of Students in Three Grades')
plt.show()
```

```python
plt.savefig('plot.png')
```
